```python
import pandas as pd

df_space = pd.read_csv('space_titanic.csv')
df_space.head()
```

```python
df_space['is_train'] = df_space['PassengerId'].str.split('_', expand=True)[0].astype(int) % 4 != 0
```

```python
from scipy.stats import chi2_contingency

for i in ['HomePlanet', 'CryoSleep', 'VIP', 'Destination', 'Transported']:
    print(i, chi2_contingency(df_space.groupby([i, 'is_train']).size().unstack())[1])
```

```python
from scipy.stats import f_oneway
for i in ['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']:
    print(i, f_oneway(*df_space.loc[df_space[i].notna()].groupby(['is_train'])[i].agg(list).tolist()))
```

```python
s_is_train = df_space.pop('is_train')

df_space.loc[s_is_train].to_csv('space_titanic_train.csv', index=False)

df_test = df_space.loc[~s_is_train].copy()

df_test.pop('Transported').to_frame().to_csv('space_titanic_ans.csv', index=False)
df_test.to_csv('space_titanic_test.csv', index=False)
```
