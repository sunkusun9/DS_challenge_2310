```python
dtype = { i : 'int'
    for i in (df_fetal - df_fetal.astype('int') == 0).mean().pipe(lambda x: x.loc[x == 1]).index.tolist()
}
```

```python
df_fetal = pd.read_csv('fetal_health.csv', dtype=dtype).rename(columns=lambda x: '_'.join(x.split(' ')))
df_fetal.to_csv('fetal_health_2.csv', index=False)
```
